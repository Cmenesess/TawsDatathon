```python
import pandas as pd
import numpy as np
import json
import datetime
```

```python
contribuyentes=pd.read_csv('../../data/ruc_info.csv',delimiter=',',na_values=[None,"NaN","null"])
```

```python
actividades=contribuyentes['informacionFechasContribuyente'].str.split(",",expand=True)
actividades
```

```python
contribuyentes['fechaInicioActividad']=actividades[0].str.split(":",expand=True)[1].replace("'",'',regex=True).str.strip().replace("None",np.nan)
contribuyentes['fechaCese']=actividades[1].str.split(":",expand=True)[1].str.strip().replace("None",np.nan).replace("'",'',regex=True)
contribuyentes['fechaInscripcion']=actividades[2].str.split(":",expand=True)[1].str.strip().replace("None",np.nan).replace("'",'',regex=True)
contribuyentes['fechaActualizacion']=actividades[3].str.split(":",expand=True)[1].str.replace("}","").str.strip().replace("None",np.nan).replace("'",'',regex=True)
```

```python
contribuyentes
```

```python
contribuyentes.isnull().sum()
```

```python
contribuyentes
```

```python
contribuyentes.drop(['numeroRucFantasma','nombreComercial','contribuyenteFantasma','informacionFechasContribuyente'],axis=1,inplace=True) 
```

```python
contribuyentes['fechaCese']=pd.to_datetime(contribuyentes['fechaCese'],errors = 'coerce',format='%d/%m/%Y')
contribuyentes['fechaInscripcion']=pd.to_datetime(contribuyentes['fechaInscripcion'],errors = 'coerce',format='%d/%m/%Y')
contribuyentes['fechaActualizacion']=pd.to_datetime(contribuyentes['fechaActualizacion'],errors = 'coerce',format='%d/%m/%Y')
contribuyentes['fechaInicioActividad']=pd.to_datetime(contribuyentes['fechaInicioActividad'],errors = 'coerce',format='%d/%m/%Y')
```

```python
contribuyentes.dropna(subset=['fechaInicioActividad'],inplace=True)
contribuyentes.isna().sum()

```
