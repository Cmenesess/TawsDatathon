```python
df["fecha_consulta"] = pd.to_datetime(df["fecha_consulta"])
```

```python
def asign_meses(x: float):
    #x = x.upper()
    if x == 1:
        return "Enero"
    if x == 2:
        return "Febrero"
    if x == 3:
        return "Marzo"
    if x == 4:
        return "Abril"
    if x == 5:
        return "Mayo"
    if x == 6:
        return "Junio"
    if x == 7:
        return "Julio"
    if x == 8:
        return "Agosto"
    if x == 9:
        return "Septiembre"
    if x == 10:
        return "Octubre"
    if x == 11:
        return "Noviembre"
    if x == 12:
        return "Diciembre"
    return x
```

```python
df["mes_fecha_inicio_teletrabajo"] = df["mes_fecha_inicio_teletrabajo"].apply(asign_meses)
```
