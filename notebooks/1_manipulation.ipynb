```python
def asign_meses(x: float):
    #x = x.upper()
    if x == 1:
        return "Enero"
    if x == 2:
        return "Febrero"
    if x == 3:
        return "Marzo"
    if x == 4:
        return "Abril"
    if x == 5:
        return "Mayo"
    if x == 6:
        return "Junio"
    if x == 7:
        return "Julio"
    if x == 8:
        return "Agosto"
    if x == 9:
        return "Septiembre"
    if x == 10:
        return "Octubre"
    if x == 11:
        return "Noviembre"
    if x == 12:
        return "Diciembre"
    return x
```

```python
import unicodedata
import string
import stanza

stanza.download('es')
nlp = stanza.Pipeline('es')
def text_preprocesing(text):
    #remove_puntuation_and_accents
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    text = text.translate(str.maketrans('', '', string.punctuation))
    #to lower
    text=text.lower()
    #remove stopwords
    from nltk.corpus import stopwords
    text = ' '.join([word for word in text.split() if word not in (stopwords.words('spanish', 'english'))])
    doc=nlp(text)
    text=' '.join([word.text for word in doc.sentences[0].words if word.upos!='PUNCT'])
    return text
```

```python
texto_unico=' '.join(lista_actividad)
```
